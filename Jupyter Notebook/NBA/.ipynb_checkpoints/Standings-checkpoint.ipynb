```python
import chart_studio.plotly as py
import plotly.graph_objs as go 
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df_east = pd.read_csv('eastern_conference.csv')
df_west = pd.read_csv('western_conference.csv')
```

```python

df_east = pd.read_csv('eastern_conference.csv')

sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(12, 8))

sns.barplot(x="Eastern Conference", y="W/L%", data=df_east, ax=ax)

ax.set_title("Eastern Conference Standings", fontsize=18)
ax.set_xlabel("Team", fontsize=14)
ax.set_ylabel("Win-Loss Percentage", fontsize=14)

plt.xticks(rotation=45, ha='right')
fig.savefig('eastern_conference_standing.png', dpi=300)

plt.show()
```

```python
df_east = pd.read_csv('western_conference.csv')

sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(12, 8))

sns.barplot(x="Western Conference", y="W/L%", data=df_east, ax=ax)

ax.set_title("Western Conference Standings", fontsize=18)
ax.set_xlabel("Team", fontsize=14)
ax.set_ylabel("Win-Loss Percentage", fontsize=14)

plt.xticks(rotation=45, ha='right')
fig.savefig('western_conference_standing.png', dpi=300)


plt.show()
```
