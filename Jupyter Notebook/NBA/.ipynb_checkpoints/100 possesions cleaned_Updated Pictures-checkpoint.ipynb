```python
import pandas as pd

# Load the CSV file into a pandas dataframe
df = pd.read_csv("NBA per 100 stats_cleaned.csv")

# Show all column names
print(df.columns)
```

# Column Descriptions
- Team: The name of the NBA team
- G: The number of games played
- MP: The number of minutes played per game
- FG: The number of field goals made per 100 possessions
- FGA: The number of field goals attempted per 100 possessions
- FG%: The field goal percentage, calculated as FG / FGA
- 3P: The number of 3-point field goals made per 100 possessions
- 3PA: The number of 3-point field goals attempted per 100 possessions
- 3P%: The 3-point field goal percentage, calculated as 3P / 3PA
- 2P: The number of 2-point field goals made per 100 possessions
- 2PA: The number of 2-point field goals attempted per 100 possessions
- 2P%: The 2-point field goal percentage, calculated as 2P / 2PA
- FT: The number of free throws made per 100 possessions
- FTA: The number of free throws attempted per 100 possessions
- FT%: The free throw percentage, calculated as FT / FTA
- ORB: The number of offensive rebounds per 100 possessions
- DRB: The number of defensive rebounds per 100 possessions
- TRB: The total number of rebounds per 100 possessions
- AST: The number of assists per 100 possessions
- STL: The number of steals per 100 possessions
- BLK: The number of blocks per 100 possessions
- TOV: The number of turnovers per 100 possessions
- PF: The number of personal fouls committed per 100 possessions
- PTS: The number of points scored per 100 possessions

```python
import pandas as pd

# Load the CSV file into a pandas dataframe
df = pd.read_csv("NBA per 100 stats.csv")

# Create a dictionary to map each team to its conference
conference_dict = {
    "Atlanta Hawks*": "Eastern",
    "Boston Celtics*": "Eastern",
    "Brooklyn Nets*": "Eastern",
    "Charlotte Hornets": "Eastern",
    "Chicago Bulls*": "Eastern",
    "Cleveland Cavaliers": "Eastern",
    "Detroit Pistons": "Eastern",
    "Indiana Pacers": "Eastern",
    "Miami Heat*": "Eastern",
    "Milwaukee Bucks*": "Eastern",
    "New York Knicks": "Eastern",
    "Orlando Magic": "Eastern",
    "Philadelphia 76ers*": "Eastern",
    "Toronto Raptors*": "Eastern",
    "Washington Wizards": "Eastern",
    "Dallas Mavericks*": "Western",
    "Denver Nuggets*": "Western",
    "Golden State Warriors*": "Western",
    "Houston Rockets": "Western",
    "Los Angeles Clippers": "Western",
    "Los Angeles Lakers": "Western",
    "Memphis Grizzlies*": "Western",
    "Minnesota Timberwolves*": "Western",
    "New Orleans Pelicans*": "Western",
    "Oklahoma City Thunder": "Western",
    "Phoenix Suns*": "Western",
    "Portland Trail Blazers": "Western",
    "Sacramento Kings": "Western",
    "San Antonio Spurs": "Western",
    "Utah Jazz*": "Western",
}

# Add a new column to the dataframe with the conference for each team
df["Conference"] = df["Team"].map(conference_dict)

# Show the first 5 rows of the dataframe with the new Conference column
print(df.head())

```

```python
df.head(40)
```

```python
df['Conference'].value_counts()
```

```python

# drop rows with missing values
df.dropna(inplace=True)

df.head(40)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# filter the DataFrame by conference
western = df[df["Conference"] == "Western"]
eastern = df[df["Conference"] == "Eastern"]

# calculate the average points per game for each conference
western_avg_pts = western["PTS"].mean()
eastern_avg_pts = eastern["PTS"].mean()

# create a bar chart
labels = ["Western", "Eastern"]
sizes = [western_avg_pts, eastern_avg_pts]
plt.bar(labels, sizes)

# add annotations for each bar
for i, size in enumerate(sizes):
    plt.annotate(round(size, 2), xy=(i, size), ha='center', va='bottom')

plt.title("Average Points per Game by Conference")
plt.ylabel("Points per Game")
plt.show()


```
