```python
# Selecting only the "Eastern Conference" and "W/L%" columns
df_east[['Eastern Conference', 'W/L%']]
```

```python
# Selecting only the teams with a W/L% greater than 0.6
df_east[df_east['W/L%'] > 0.6]

# Selecting only the teams that made it to the playoffs (marked with *)
df_east[df_east['Eastern Conference'].str.endswith('*')]
```

```python
# Sorting the teams by their win-loss record in descending order
df_east.sort_values(by='W', ascending=False)

# Sorting the teams by their SRS rating in ascending order
df_east.sort_values(by='SRS')
```

```python
# Sorting the teams by their win-loss record in descending order
df_east.sort_values(by='W', ascending=False)

# Sorting the teams by their SRS rating in ascending order
df_east.sort_values(by='SRS')
```

```python
# Getting the average points scored by all teams in the conference
df_east['PS/G'].mean()

# Getting the maximum and minimum win totals in the conference
df_east['W'].max(), df_east['W'].min()

# Grouping the teams by their W/L% and getting the average SRS rating for each group
df_east.groupby('W/L%')['SRS'].mean()
```

```python
# Creating a new column that calculates the point differential for each team
df_east['Point Differential'] = df_east['PS/G'] - df_east['PA/G']
df_east
```

```python
# Creating a new column that indicates whether a team made it to the playoffs or not
df_east['Playoff Team'] = df_east['Eastern Conference'].str.endswith('*')
df_east
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot of the playoff and non-playoff teams
sns.countplot(x='Playoff Team', data=df_east)

# Add labels and title to the plot
plt.xlabel('Playoff Team')
plt.ylabel('Count')
plt.title('Number of Eastern Conference Playoff Teams')

# Display the plot
plt.show()

```
