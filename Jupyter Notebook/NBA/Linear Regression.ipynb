```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# load the data
df = pd.read_csv("NBA per 100 stats_cleaned.csv")

# remove non-numeric columns
non_numeric_cols = ['Team']
df_numeric = df.drop(columns=non_numeric_cols)

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    df_numeric.drop(columns=["PTS"]),
    df_numeric["PTS"],
    test_size=0.2,
    random_state=42,
)

# impute missing values with mean
imputer = SimpleImputer(strategy="mean")
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

# Train a linear regression model on the training data
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Evaluate the model
score = model.score(X_test, y_test)
print(f"R^2 Score: {score}")
```

## Training and Testing Data

```python
X = df_cleaned[['MP', 'FG', 'FGA', '3P', '3PA', '2P', '2PA', 'FT', 'FTA', 'ORB', 'DRB', 'AST', 'STL', 'BLK', 'TOV', 'PF']]
y = df_cleaned['PTS']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print(lm.coef_)
```

## Predicting Test Data

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('True PTS')
plt.ylabel('Predicted PTS')
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, predictions)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```

## Evaluating the Model

```python
import numpy as np
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

When looking at the coefficients, the focus should be on their magnitude and sign. A larger magnitude indicates a stronger relationship between the independent variable and the dependent variable. A positive sign indicates a positive relationship, meaning that an increase in the independent variable will lead to an increase in the dependent variable, while a negative sign indicates a negative relationship, meaning that an increase in the independent variable will lead to a decrease in the dependent variable. The coefficients can be used to interpret the impact of each independent variable on the dependent variable.
