W: Number of wins
L: Number of losses
W/L%: Win-loss percentage, calculated as W / (W + L)
GB: Games behind the first place team in the conference
PS/G: Points scored per game
PA/G: Points allowed per game
SRS: Simple Rating System, a rating that takes into account average point differential and strength of schedule. A positive SRS indicates that a team is above average, and a negative SRS indicates that a team is below average.
