# Column Descriptions
- Team: The name of the NBA team
- G: The number of games played
- MP: The number of minutes played per game
- FG: The number of field goals made per 100 possessions
- FGA: The number of field goals attempted per 100 possessions
- FG%: The field goal percentage, calculated as FG / FGA
- 3P: The number of 3-point field goals made per 100 possessions
- 3PA: The number of 3-point field goals attempted per 100 possessions
- 3P%: The 3-point field goal percentage, calculated as 3P / 3PA
- 2P: The number of 2-point field goals made per 100 possessions
- 2PA: The number of 2-point field goals attempted per 100 possessions
- 2P%: The 2-point field goal percentage, calculated as 2P / 2PA
- FT: The number of free throws made per 100 possessions
- FTA: The number of free throws attempted per 100 possessions
- FT%: The free throw percentage, calculated as FT / FTA
- ORB: The number of offensive rebounds per 100 possessions
- DRB: The number of defensive rebounds per 100 possessions
- TRB: The total number of rebounds per 100 possessions
- AST: The number of assists per 100 possessions
- STL: The number of steals per 100 possessions
- BLK: The number of blocks per 100 possessions
- TOV: The number of turnovers per 100 possessions
- PF: The number of personal fouls committed per 100 possessions
- PTS: The number of points scored per 100 possessions

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the email dataset into a pandas dataframe
df = pd.read_csv("emails.csv")

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df["text"], df["spam"], test_size=0.2, random_state=42)

# Convert the text data to a bag of words representation
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)

# Train a Naive Bayes classifier on the training data
clf = MultinomialNB()
clf.fit(X_train_vectorized, y_train)

# Vectorize the testing data and make predictions
X_test_vectorized = vectorizer.transform(X_test)
y_pred = clf.predict(X_test_vectorized)

# Evaluate the model's accuracy on the testing data
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```
