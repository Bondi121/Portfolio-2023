```python
import pandas as pd

# Load the NBA playoff data
eliminated_first_round = ['Brooklyn Nets', 'Atlanta Hawks', 'Chicago Bulls', 'Toronto Raptors', 'Utah Jazz', 'Denver Nuggets', 'Minnesota Timberwolves', 'New Orleans Pelicans']
eliminated_second_round = ['Milwaukee Bucks', 'Philadelphia 76ers', 'Phoenix Suns', 'Memphis Grizzlies']
eliminated_third_round = ['Miami Heat', 'Dallas Mavericks']
eliminated_fourth_round = ['Boston Celtics']
champion = ['Golden State Warriors']
did_not_make_playoffs = ['Charlotte Hornets', 'Sacramento Kings', 'Orlando Magic', 'Portland Trail Blazers', 'Houston Rockets', 'Los Angeles Lakers', 'Indiana Pacers', 'San Antonio Spurs', 'Detroit Pistons', 'New York Knicks', 'Los Angeles Clippers', 'Cleveland Cavaliers', 'Washington Wizards', 'Oklahoma City Thunder']

data1 = {'Team': eliminated_first_round + eliminated_second_round + eliminated_third_round + eliminated_fourth_round + champion + did_not_make_playoffs,
         'Season Success': ['Eliminated in First Round'] * len(eliminated_first_round) + ['Eliminated in Second Round'] * len(eliminated_second_round) + ['Eliminated in Third Round'] * len(eliminated_third_round) + ['Eliminated in Fourth Round'] * len(eliminated_fourth_round) + ['Champion'] * len(champion) + ['Did Not Make Playoffs'] * len(did_not_make_playoffs)}

nba_playoff_df = pd.DataFrame(data1)

# Load the NBA per 100 stats data
nba_stats_df = pd.read_csv('NBA per 100 stats_cleaned.csv')

# Remove the * from the team names
nba_stats_df['Team'] = nba_stats_df['Team'].str.rstrip('*')

# Merge the two dataframes
merged_df = pd.merge(nba_playoff_df, nba_stats_df, on='Team')

# Rename the 'G' column to 'Games Played'
merged_df = merged_df.rename(columns={'G': 'Games Played'})

# Add the 'Season Success' column
merged_df['Season Success'] = merged_df['Season Success'].fillna('Playoffs')

# Show the merged dataframe
merged_df.head(40)
```

```python
sorted_df = merged_df.sort_values(by='Season Success')
sorted_df
```

```python
merged_df['Season Success'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

order = ['Did Not Make Playoffs', 'Eliminated in First Round', 'Eliminated in Second Round',
         'Eliminated in Third Round', 'Eliminated in Fourth Round', 'Champion']

sns.barplot(x='Season Success', y='PTS', data=merged_df, order=order, ci=None)
plt.title('Mean 3P% by Season Success')
plt.xticks(rotation=75)
plt.ylabel('Mean 3-Point Percentage')

# Make the x-label and y-label bold
font = font_manager.FontProperties(weight='bold')
plt.xlabel('Season Success', fontproperties=font)
plt.ylabel('Mean 3-Point Percentage', fontproperties=font)

plt.show()
```
