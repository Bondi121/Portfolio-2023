```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df_east = pd.read_csv('eastern_conference.csv')

# Set style
sns.set(style="whitegrid")

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Create horizontal bar plot
sns.barplot(x="W/L%", y="Eastern Conference", data=df_east, ax=ax)

# Set title and axis labels
ax.set_title("Eastern Conference Standings", fontsize=18)
ax.set_xlabel("Win-Loss Percentage", fontsize=14)
ax.set_ylabel("Team", fontsize=14)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# Save figure
plt.savefig('eastern_standings.png', bbox_inches='tight')

# Show plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df_east = pd.read_csv('eastern_conference.csv')

# Create correlation matrix
corr_matrix = df_east[['W/L%', 'PS/G']].corr()

# Set style
sns.set(style="white")

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu", ax=ax)

# Set title
ax.set_title("Eastern Conference Correlation Matrix (W/L% vs. PS/G)", fontsize=18)

# Save figure
plt.savefig('eastern_corr_matrix.png', bbox_inches='tight')

# Show plot
plt.show()
```
