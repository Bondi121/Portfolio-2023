```python
from newsapi import NewsApiClient

def get_news_sources(api_key):
    # Create an instance of NewsApiClient using the provided API key
    newsapi = NewsApiClient(api_key=api_key)

    # Call the get_sources() method to fetch all the news sources from the API
    response = newsapi.get_sources()

    # Extract the list of sources from the response dictionary
    sources = response['sources']

    # Create a list of source IDs by iterating over the sources
    source_ids = [source['id'] for source in sources]

    # Truncate the list to a size of 10 using slicing
    source_ids = source_ids[:10]

    # Return the truncated list of source IDs
    return source_ids

# Call the get_news_sources function with your API key
api_key = '1ea444be81d141d5a5ec72566c47501b'
source_ids = get_news_sources(api_key)

# Print the truncated list of source IDs
print(source_ids)

# Create an empty list to store the news articles
data_list = []

# Iterate through each source ID
for source_id in source_ids:
    # Call the get_everything() method to fetch news from the current source ID
    response = newsapi.get_everything(sources=source_id)

    # Extract the list of articles from the response
    articles = response['articles']

    # Add the articles to the data_list
    data_list.extend(articles)

# Print the total number of news articles retrieved
print('Total News:', len(data_list))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Split the data into training and testing sets
training_x, testing_x, training_y, testing_y = train_test_split(df['text'], df['label'], test_size=0.3, random_state=7)

# Fill NaN values with empty strings
training_x = training_x.fillna('')
testing_x = testing_x.fillna('')

# Create an instance of CountVectorizer
count_vectorizer = CountVectorizer(stop_words='english', max_df=0.7)

# Fit and transform the training text data
feature_train = count_vectorizer.fit_transform(training_x)

# Transform the testing text data
feature_test = count_vectorizer.transform(testing_x)

# Replace NaN values in training labels with "Unknown"
training_y = training_y.fillna("Unknown")

# Initialize the classifier
classifier = PassiveAggressiveClassifier(max_iter=50)

# Fit the classifier to the training data
classifier.fit(feature_train, training_y)

# Predict the labels for the testing data
prediction = classifier.predict(feature_test)

# Calculate the accuracy score
score = accuracy_score(testing_y.astype(str), prediction.astype(str))
print("Accuracy: ", score * 100)
```
